## Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
**Answer:**
A **Class** is a blueprint or template used to create objects. It defines properties (variables) and behaviors (methods).
An **Object** is an instance of a class. It represents a real-world entity and can use the properties and methods defined in the class.


```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print('Name:', self.name)
        print('Age:', self.age)

# Creating object
s1 = Student('Rahul', 20)
s1.display()
```

## Q3. Explain why the __init__() function is used. Give a suitable example.
**Answer:**
The **__init__()** function is a constructor in Python. It is automatically called when an object of a class is created. It is used to initialize object properties.


```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def show(self):
        print(self.brand, self.model)

c1 = Car('Toyota', 'Fortuner')
c1.show()
```

## Q4. Why self is used in OOPs?
**Answer:**
`self` is used to refer to the current object of the class. It allows access to the variables and methods of the class within the class methods.


```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print('Hello', self.name)

p = Person('Aman')
p.greet()
```

## Q5. What is inheritance? Give an example for each type of inheritance.
**Answer:**
Inheritance is a feature of OOP where one class can inherit properties and methods from another class.

**Types of inheritance:**
1. Single Inheritance
2. Multiple Inheritance
3. Multilevel Inheritance
4. Hierarchical Inheritance


```python
# Single Inheritance
class Animal:
    def speak(self):
        print('Animal makes sound')

class Dog(Animal):
    def bark(self):
        print('Dog barks')

d = Dog()
d.speak()
d.bark()

# Multiple Inheritance
class Father:
    def skills1(self):
        print('Gardening')

class Mother:
    def skills2(self):
        print('Cooking')

class Child(Father, Mother):
    pass

c = Child()
c.skills1()
c.skills2()

# Multilevel Inheritance
class A:
    def showA(self):
        print('Class A')

class B(A):
    def showB(self):
        print('Class B')

class C(B):
    def showC(self):
        print('Class C')

obj = C()
obj.showA()
obj.showB()
obj.showC()

# Hierarchical Inheritance
class Parent:
    def display(self):
        print('Parent class')

class Child1(Parent):
    pass

class Child2(Parent):
    pass

c1 = Child1()
c2 = Child2()

c1.display()
c2.display()
```
